### Index and Filtering

```python
auto_csv = r"C:\Users\ramananb\Downloads\Automobile.csv"
```

```python
import pandas as pd
df = pd.read_csv(auto_csv)
df
```

```python
"""
iloc - used to access a portion of a dataframe (rows & cols - using integers/positions)
Note: If there is no custom index set for rows, then by default the index is set from 0 
and the subsequent indices have values incremented by 1
"""
df.iloc[1:4, 3:6]   # accessing the rows through index 1 to 3 and columns through index 3 to 5
```

```python
df.iloc[[3,6, 9, 11, 100], :] # accessing different row indices without having an actual range
```

```python
df.iloc[[3,6, 9, 11, 100], [0, 2, 6]] # accessing different row and column indices without having an actual range
```

```python
# Set custom index
df.set_index('name', inplace=True)  
# inplace=True will actually update the dataframe, without the argument df will not be impacted
```

```python
df
```

```python
# access the row indices with labels (ie., custom index values)
df.loc[['plymouth satellite','ford mustang gl','dodge rampage']] 
```

```python
# accessing the rows and columns with actual labels
df.loc[['plymouth satellite','ford mustang gl','dodge rampage'], ['mpg','cylinders','horsepower']]
```

```python
# accessing row and column labels by having them in list variables
idx = ['plymouth satellite','ford mustang gl','dodge rampage','chevy s-10','vw pickup']
cols = ['mpg','cylinders','horsepower', 'origin']
df.loc[idx, cols]
```

```python
df
```

```python
# Column value to list
mpg_lst = df['mpg'].to_list()
mpg_lst
```

```python
# Index values to list
idx_lst = list(df.index)  # can also use df.index.values
idx_lst
```

### Filtering the Dataframe using conditions

```python
# for example we want to filter the data where mpg >=18
df.loc[df['mpg']>=18]   # we can also store this sliced dataframe into another dataframe by assigning it to a variable
```

```python
df.loc[df['mpg']>=18, :]
```

```python
df.loc[df['mpg']>=18, ['mpg', 'weight', 'origin']]
```

### Dataframe filter with multiple conditions using logical operators

```python
# mpg >=18 and origin=='usa'
df.loc[(df['mpg']>=18)&(df['origin']=='usa')]
```

```python
# mpg >=18 or origin=='usa'
df.loc[(df['mpg']>=18)|(df['origin']=='usa')]
```

### we can also access the filters by assigning them to variables

```python
filter1 = ((df['mpg']>=18)|(df['origin']=='usa'))
df.loc[filter1]
```

```python
# Resetting the index
df.reset_index(inplace=True)
```

### Filtering based on multi-index
we use list of tuples

```python
df.loc[[('chevrolet impala', 14.0),('amc gremlin', 19.0)]]
```

```python
df.loc[['chevrolet impala','amc gremlin']]
```

```python
df
```

### Filtering based on multi-index but skipping level 0 index and using only level 1 index

```python
df[df.index.isin([14], level=1)]  
# Note: 14 is not a position/index but index value for mpg 
```

```python
df[df.index.isin([14, 17, 32], level=1)]  # we can also pass list of index values
```

```python
df[df.index.isin([14.0, 17, 32], level=1)]
```

```python
mpg_lst = [14.0, 17, 32]
df[df.index.isin(mpg_lst, level=1)]   # we can also pass a list variable
```

### Another way of filtering based on multi-index but skipping level 0 index and using only level 1 index

```python
df.loc[(slice(None),14),:] 
# in this method we have to explicitly specify even if we want all the columns unlike the previous methods
```
