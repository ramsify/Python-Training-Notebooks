```python
import pandas as pd
import numpy as np

imdb_file = r"C:\Users\ramananb\Downloads\Horror Movies IMDb.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df
```

```python
dict1 = {'Title': ['Saw 2','Saw 3'],
        'Year': ['2005','2006'],
        'Runtime': [105, 108],
        'Genre': ['Horror','Thriller'],
        'Rating': [8,7],
        'Votes': [435678,435679],
        'Gross': ['$100M','$200M']}
df1 = pd.DataFrame(dict1)
df1
```

### Let's have another dataframe with just Movie Title and Production Studio

```python
# Include another column - Production Studio
dict3 = {
        'Title': ['Saw', 'Saw 2', 'Saw 3', 'Hostel','Hostel 2','Hostel 3'],
        'Production Studio': 'Lionsgate',
         }
df3 = pd.DataFrame(dict3)
df3
```

```python
# Include another column - Production Studio
dict3 = {
        'Title': ['Saw', 'Saw 2', 'Saw 3', 'Hostel','Hostel 2','Hostel 3'],
        'Production Studio': ['Lionsgate']*3 + ['Sony Picutres']*3,
         }
df3 = pd.DataFrame(dict3)
df3
```

### Concating df1, df2 and df3 along the column axis

```python
pd.concat([df1,df2,df3],axis=1)
```

### Let's first stack the df1 and df2 along the row axis to make them into 1 dataframe

```python
merged_df = pd.concat([df1,df2],ignore_index=True)
merged_df
```

### Concatenating the merged_df and df3 along the column axis

```python
pd.concat([merged_df,df3],axis=1)
```

```python
# merge() function - used to merge two dataframes based on the specified columns either as an inner/outer/left/right join
merged_df.merge(df3)
```

```python
merged_df.merge(df3, how='inner')
```

```python
merged_df.merge(df3, how='outer')
```

```python
df3
```

```python
# Always ensure to specify the column name on which the join has to happen
# If the column names on which the join has to happen are same for the two dfs then just use the 'on' parameter
# If you want to join on more than 1 column just pass the column labels as a list to the 'on' parameter
merged_df.merge(df3, how='outer', on='Title')
```

```python
df3.rename(columns={'Title':'Movie Title'},inplace=True)
df3
```

### Now we have two different column names on which the merge has to happen on the 2 dfs 
#### If the column names are different then use left_on , right_on and the corresponding column names

```python
merged_df2 = merged_df.merge(df3, how='outer', left_on='Title', right_on='Movie Title')
merged_df2
```

```python
merged_df2.drop(columns=['Movie Title'],inplace=True)
merged_df2
```

### For merging more than 2 dataframes you can use either of the below ways
#### 1) pd.merge in a nested structure
#### pd.merge(pd.merge(df1,df2, how= join condition>, on = column values),df3,how= join condition, on = column values)
#### 2) chain the merge() functions
#### df1.merge(df2, how= join condition>, on = column values).merge(df3,how= join condition, on = column values)
