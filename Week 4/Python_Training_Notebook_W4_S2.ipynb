```python
# Normal function
def cust_print(cntry, cap):
    print('{} is the capital of {}'.format(cap, cntry))
    
cust_print('Germany', 'Berlin')
```

```python
def cust_print(cntry, cap):
    print('{1} is the capital of {0}'.format(cntry, cap))
    
cust_print('Germany', 'Berlin')
```

### Lambda - anonymous functions - they don't have a proper identity or name
### it can have any number of arguments but can have only single expression unlike normal functions

```python
cntry_cap = lambda tup:   print('{1} is the capital of {0}'.format(*tup))
tup = ('Austria', 'Vienna')
cntry_cap(tup)
```

### In the above column if you observe the values, '$' and 'M' values are redundant
### Let's remove them and store it in a new column and convert the dtype = float so we can do statistical or arithmetic operations

```python
# replacing $ and 'M' values using the str.replace chain method
imdb_df['Gross in million $'] = imdb_df['Gross'].fillna('0').apply(lambda x: x.replace('$','').replace('M','')) 
```

```python
# converting the column to float
imdb_df['Gross in million $'] = imdb_df['Gross in million $'].astype(float)
```

```python
# Storing the mean/average value in a variable
avg = imdb_df['Gross in million $'].mean()
avg, type(avg)
```

### Creating a new column which has the below logic:
0 - if the 'Gross in million' value is < avg    
1 - if 'Gross in million' value' is >= avg

```python
imdb_df['Movie Category based on Avg Gross'] = imdb_df['Gross in million $'].apply(lambda x: 0 if(x<avg) else 1) 
```
