# Datacompy - third party library to compare dataframes for any discrepancies
To understand how datacompy works:
1) Lets create a dataframe out of an API
2) Make a copy of the dataframe and store it in another variable
3) Make some random edits to the second dataframe and compare both the dfs

To know more about datacompy library visit - https://capitalone.github.io/datacompy/


```python
import requests
import json
import numpy as np
url = "https://imdb-top-100-movies.p.rapidapi.com/"
headers = {
"X-RapidAPI-Key": "a85d7f29f6msh0221e41e28e2b71p19ab87jsnfab3ce8fa497",
"X-RapidAPI-Host": "imdb-top-100-movies.p.rapidapi.com"
}
response = requests.get(url, headers=headers)
json_txt = response.json()
json_txt
```

```python
import pandas as pd
movie_df = pd.json_normalize(json_txt)
movie_df
```

```python
movie_df2 = movie_df.copy()
movie_df2
```

```python
# We are just updating the 'rating' and 'id' column
movie_df2.loc[movie_df2['title']=='The Godfather',['rating','id']] = ['9.8','topp']
```

```python
# we are also adding 1 more row to movie_df2 
movie_df2.loc[len(movie_df2)] = [11, 'Behind Enemy Lines',np.nan,9.4,np.nan,2001,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
```

```python
movie_df2[movie_df2['title']=='Behind Enemy Lines']
```

### Using the datacompy library for comparing the 2 dataframes

```python
import datacompy
comparison = datacompy.Compare(
movie_df,
movie_df2,
join_columns='title',
df1_name='Source',
df2_name='Target'
)
print(comparison.report())
```

### Hierarchical Indexing

```python
for df in movie_df,movie_df2:
    df.set_index('title',inplace=True)
```

```python
movie_df.head(3)
```

```python
movie_df2.head(3)
```

```python
# We can use concat function with Keys parameters to assign another level of column indexing to the dataframes
# With these labels while concatenating the dfs across column axis we can tell apart the data from each df
df = pd.concat([movie_df,movie_df2],keys=['Source','Target'],axis=1)
df
```

```python
# Filtering a row from a dataframe having hierarchical index
df[df['Source']['rank']==2]
```
